```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'pah_2016_2022_with_Y_M_02.csv', encoding='latin1')
```

## Yearly data analysis

```python
df.groupby(["yyyy"], as_index=True).agg(count=("yyyy", "count"))
```

```python
df_yearly = df.groupby(["yyyy"], as_index=True).agg(count=("yyyy", "count"))
```

```python
df_yearly.plot(kind="line", y="count", legend=None)
# set the axis labels and title
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("2016_2022 Yearly Record")
# show the plot
plt.show()
```

```python
df_yearly.plot(kind="bar", y="count", legend=None)
# set the axis labels and title
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("2016_2022 Yearly Record")
# show the plot
plt.show()
```

```python
df_yearly.plot(kind="pie", y="count", legend=None)
# set the axis labels and title
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("2016_2022 Yearly Record")
# show the plot
plt.show()
```
